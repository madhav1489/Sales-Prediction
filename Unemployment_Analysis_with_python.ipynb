```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
df=pd.read_csv('/content/Unemployment in India.csv')
```

```python
df.head(30)
```

```python
df.describe()
```

```python
df.columns
```

```python
#removing whitespace from column names
df.columns=df.columns.str.strip()
df.columns
```

```python
#checking for missing values
df.isnull().sum()
```

```python
#dropping records with missing values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
#checking for duplicate records
df.duplicated().sum()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('Correlation heatmap for unemployment in India (20192020)', fontsize=20)
```

```python
sns.pairplot(df)
```

```python
df['Area'].value_counts()
```

```python
plt.pie(df['Area'].value_counts(), labels=df['Area'].value_counts().index, autopct='%1.1f%%')
```

```python
avg_unemp_rate = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(12,6))
sns.barplot(x=avg_unemp_rate.index, y=avg_unemp_rate.values)

plt.title('Region vs Average Unemployment Rate\n20192020', fontsize=20)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=90)

plt.show()

print(f"{avg_unemp_rate.idxmax()} had the highest average unemployment rate of {avg_unemp_rate.max():.2f}%")
print(f"{avg_unemp_rate.idxmin()} had the lowest average unemployment rate of {avg_unemp_rate.min():.2f}%")
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)',hue='Region',data=df)
plt.title('Variation of Estimated Unemployment Rate in different regions over time\n20192020', fontsize=20)
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
```

```python
plt.figure(figsize=(12,6))
sns.set()
sns.scatterplot(x='Estimated Employed', y='Estimated Unemployment Rate (%)',hue='Region',data=df)
plt.title('Estimate Labour Participation Rate (%) vs Estimated Unemployment Rate(%)\n20192020', fontsize=20)
plt.xlabel('Estimated Employed')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
```

```python
plt.figure(figsize=(12,6))
sns.set()
sns.scatterplot(x='Estimated Labour Participation Rate (%)',y='Estimated Unemployment Rate (%)',hue='Region',data=df)
plt.title('Estimated Labour Participation Rate (%) vs Estimated Unemployment Rate (%)\n20192020',fontsize=20)
plt.xlabel('Estimated Labour Participation Rate (%)')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.legend(loc='center left', bbox_to_anchor=(1.1, 0.5))
```
