Model Selection and Evaluation

```python
X=df.drop(columns=['Sales'],axis =1).values
y=df['Sales'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=5)
```

```python
from re import L
from sklearn.metrics import mean_squared_error, r2_score
#Implementing Lasso Regression
from sklearn.linear_model import Lasso
ls=Lasso()
ls.fit(X_train,y_train)

y_pred=ls.predict(X_test)
print("Mean Sqaured Error:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```

```python
#Implementing Ridge Regressor
from sklearn.linear_model import Ridge
rd=Ridge()
rd.fit(X_train,y_train)

y_pred=rd.predict(X_test)
print("Mean Sqaured Error:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```

```python
#Implementing multiple Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)

y_pred=lr.predict(X_test)
print("Mean Sqaured Error:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```

```python
#Implementing RandomForest Regressor
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(X_train,y_train)

y_pred=rf.predict(X_test)
print("Mean Sqaured Error:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```

```python
#Implementing DecisionTree Regressor
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(X_train,y_train)

y_pred=dt.predict(X_test)
print("Mean Sqaured Error:",mean_squared_error(y_test,y_pred))
print("R2 Score:",r2_score(y_test,y_pred))
```

#Conclusion
We select RandomForest Regressor as our sales prediction model as it shows consistent good r2 score and low mean squared error over most variations of the dateset.
